# CRIMEAN WAR PROJECT 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path_1854 = r"C:/Users/david/Downloads/Data Science SCA/mini project/Nightingale/raw_night_1854_1855.xlsx"
df_1854 = pd.read_excel(file_path_1854)

# 1. Display the first few rows to show data structure
print("--- First 5 Rows ---")
print(df_1854.head())

# 2. Check data types and identify variables 
print("\n--- Data Information (Types and Variables) ---")
print(df_1854.info())

# 3. Provide simple summary statistics
print("\n--- Summary Statistics ---")
print(df_1854.describe())
```

```python
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

# Define the data
months = df_1854['Month']
n_months = len(months)
angles = np.linspace(0, 2 * np.pi, n_months, endpoint=False)
width = (2 * np.pi) / n_months

# Calculate radii (Square root ensures the area is proportional to the data) [cite: 36]
disease_radius = np.sqrt(df_1854['Disease Deaths'])
wounds_radius = np.sqrt(df_1854['Wounds'])
other_radius = np.sqrt(df_1854['Other Causes'])

# Plot the wedges (Blue for Disease, Red for Wounds, Black for Other) 
ax.bar(angles, disease_radius, width=width, color='skyblue', label='Preventable Disease', alpha=0.7, edgecolor='white')
ax.bar(angles, other_radius, width=width, color='black', label='Other Causes', alpha=0.8)
ax.bar(angles, wounds_radius, width=width, color='red', label='Wounds', alpha=0.6)

# Formatting the chart
ax.set_theta_zero_location('N')  # Set January/Start to the top
ax.set_theta_direction(-1)       # Go clockwise
ax.set_xticks(angles)
ax.set_xticklabels(months)
ax.set_yticklabels([])           # Hide the radius numbers for a cleaner look
plt.title("Causes of Mortality in the Army in the East\nApril 1854 to March 1855", fontsize=15, pad=20)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

```python
# the polar plot
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

# Define the data
months = df_1855['Month']
n_months = len(months)
angles = np.linspace(0, 2 * np.pi, n_months, endpoint=False)
width = (2 * np.pi) / n_months

# Calculate radii (Square root ensures area proportionality)
# Note: use the columns from the 1855-1856 dataframe
disease_radius = np.sqrt(df_1855['Disease Deaths'])
wounds_radius = np.sqrt(df_1855['Wounds'])
other_radius = np.sqrt(df_1855['Other Causes'])

# Plot the wedges
# Plot Disease first (largest), then Other, then Wounds (smallest)
ax.bar(angles, disease_radius, width=width, color='skyblue', label='Preventable Disease', alpha=0.7, edgecolor='white')
ax.bar(angles, other_radius, width=width, color='black', label='Other Causes', alpha=0.8)
ax.bar(angles, wounds_radius, width=width, color='red', label='Wounds', alpha=0.6)

# Formatting
ax.set_theta_zero_location('N') 
ax.set_theta_direction(-1)      
ax.set_xticks(angles)
ax.set_xticklabels(months)
ax.set_yticklabels([]) # Removes the radial grid numbers

plt.title("Causes of Mortality in the Army in the East\nApril 1855 to March 1856", fontsize=15, pad=20)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()

#In this second period (1855-1856), the overall size of the wedges is significantly smaller than in the previous year. 
#Specifically,the sky-blue wedges (preventable diseases) shrank drastically after March 1855, 
#which coincides with the arrival of the Sanitary Commission and improvements in hospital hygiene.
```

```python
import matplotlib.pyplot as plt

#comparing the total deaths by category for both years
totals_54 = [df_1854['Disease Deaths'].sum(), df_1854['Wounds'].sum(), df_1854['Other Causes'].sum()]
totals_55 = [df_1855['Disease Deaths'].sum(), df_1855['Wounds'].sum(), df_1855['Other Causes'].sum()]

categories = ['Disease', 'Wounds', 'Other']
x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, totals_54, width, label='1854-1855', color='steelblue')
rects2 = ax.bar(x + width/2, totals_55, width, label='1855-1856', color='orange')

# Add labels and title
ax.set_ylabel('Total Number of Deaths')
ax.set_title('Total Mortality Comparison: Before vs. After Sanitary Reforms')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Adding data labels on top of bars
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.show()
```

# Project 1 Reflection: 
This analysis of the Crimean War mortality data highlights the transformative power of data visualization in public health. By reconstructing Florence Nightingale’s famous 'Coxcomb' or Polar Area charts, the data clearly reveals a startling reality: the vast majority of military deaths were not caused by battlefield wounds, but by preventable 'Zymotic' diseases due to poor hospital sanitation.

In the first period (1854-1855), the massive blue wedges representing disease dwarf the red wedges for wounds. However, the second period (1855-1856) shows a dramatic reduction in these blue wedges, coinciding precisely with the arrival of the Sanitary Commission and the implementation of better hygiene practices.

This project demonstrates that data is most effective when it is communicated clearly. Nightingale’s decision to use a visual format rather than a dense table of numbers allowed her to prove to the British Parliament that hospital reform was a matter of life and death. Modern data science follows this same principle—using visualization to uncover hidden patterns and drive evidence-based decision-making. Through this project, I have learned that cleaning and visualizing historical data can provide timeless lessons on the importance of infrastructure and systemic reform.

# Telecom Customer Churn PROJECT 2

```python
import pandas as pd

churn_path = r"C:/Users/david/Downloads/Data Science SCA/mini project/telco/telco_customer_churn2.csv"

df_churn = pd.read_csv(churn_path)

print("--- Customer Churn Data Head ---")
print(df_churn.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Rename the specific columns needed for the project
# match exact messy names to clean ones
clean_names = {
    'c!ontract': 'Contract',
    'ch_u+rn ': 'Churn', # Note the space at the end!
    'tenure': 'Tenure',
    'mon^thly_char^ges': 'MonthlyCharges',
    'TotalCharges': 'TotalCharges'
}

df_churn = df_churn.rename(columns=clean_names)

# 2. Final numeric fix for TotalCharges (just in case)
df_churn['TotalCharges'] = pd.to_numeric(df_churn['TotalCharges'], errors='coerce')

print("Columns successfully cleaned!")
print(df_churn[['Contract', 'Churn', 'Tenure', 'MonthlyCharges', 'TotalCharges']].head())
```

```python
print(df_churn.info())
display(df_churn.describe())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df_churn, x='Contract', hue='Churn', palette='viridis')
plt.title('Visualization 1: Churn Count by Contract Type')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_churn, x='Churn', y='Tenure', palette='magma')
plt.title('Visualization 2: How Tenure (Months) Impacts Churn')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df_churn, x='MonthlyCharges', hue='Churn', fill=True, palette='crest')
plt.title('Visualization 3: Distribution of Monthly Charges by Churn')
plt.show()
```

# Final Project Reflection
This analysis of the Telco Customer Churn dataset demonstrates the vital role of data preprocessing and cleaning in the data science workflow. Upon initial inspection, the dataset contained highly corrupted column headers, such as 'c!ontract' and 'ch_u+rn ', and numerical data stored as strings. By systematically renaming these variables and converting the 'TotalCharges' column into a numeric format, I was able to transform raw, unusable data into a clean structure suitable for visualization.

The resulting charts reveal three critical insights into customer behavior. First, the Contract Type analysis shows that 'Month-to-month' customers are at the highest risk of leaving, suggesting that long-term commitments are a primary driver of retention. Second, the Tenure box plot indicates that churn is most frequent during the first year of service, highlighting a need for better early-stage customer engagement. Finally, the Monthly Charges distribution suggests that higher costs correlate with increased churn.

To reduce attrition, the company should focus on incentivizing annual contracts and providing loyalty discounts to new users. This project illustrates how data visualization can turn a chaotic dataset into a clear, actionable strategy for improving business sustainability and customer loyalty.
